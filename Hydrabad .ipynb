 My objective is to classify 150 words based on low risk cluster 1 to high risk cluster 4. Here new_df dataframe target varaible is ward_lgd_n . first indepent varaible is elv_mean	thereshold value is  >=581, 551-580, 501-550, <=500. Second I use varible  buchg_per here I set threshold. Thrid varaibale rain_mean threshold value are below 20000, 20000-20500, 20600-21000, above 21000. Forth vaiable is  slp_mean threshold are 1.78-2.50, 2.51-3.50, 3.51-4.50, 4.51-5.41. Using this information make random forest classification model make one column cluster based on result put ward_lgd_n identify fall in which cluster

```python
# Selecting specific columns
df = new_df[['ward_lgd_n', 'buchg_per', 'slp_mean', 'elv_mean','rain_mean']].copy()
```

```python
df.head()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
from IPython.display import FileLink

# Assuming 'df' is your DataFrame and it has been preprocessed

# Define independent variables (X) and target variable (y)
X = df[['slp_mean', 'rain_mean', 'buchg_per', 'elv_mean']]
y = df['ward_lgd_n']

# Train RandomForestClassifier model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X, y)

# Make predictions with RandomForestClassifier
df['rf_predictions'] = rf_model.predict(X)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['kmeans_cluster'] = kmeans.fit_predict(X)

# Map kmeans_cluster to cluster names (assuming this is your intention)
cluster_mapping = {
    0: 'Low Risk',
    1: 'Medium Risk',
    2: 'High Risk',
    3: 'Extreme Risk'
}
df['cluster_name'] = df['kmeans_cluster'].map(cluster_mapping)

# Save the DataFrame to a CSV file
output_file = 'cluster_predictions.csv'
df.to_csv(output_file, index=False)

# Create a download link for the CSV file
csv_link = FileLink(output_file)

# Display the download link
csv_link
```

This approach combines supervised and unsupervised learning techniques to classify municipalities by flood risk, providing a robust and interpretable model. The RandomForestClassifier gives us a prediction based on known data, while KMeans clustering helps to uncover natural groupings within the data that align with different risk levels. This combined methodology ensures that we can make well-informed predictions and understand the underlying patterns in the data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Sort the DataFrame by 'rain_mean' in descending order and get the top 10
top_10_df = df.sort_values(by='rain_mean', ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='rain_mean', y='ward_lgd_n', data=top_10_df, orient='h')
plt.title('Top 10 Wards based on Mean Rainfall')
plt.xlabel('Mean Rainfall')
plt.ylabel('Ward')
plt.show()
```

```python
# Step 3: Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['rain_mean'], marker='o', linestyle='-', color='b')

# Adding title and labels
plt.title('Distribution of Rain Mean')
plt.xlabel('Index')
plt.ylabel('Rain Mean')

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
# Sort the DataFrame by 'rain_mean' and select the lowest 10 rows
df_sorted = df.sort_values(by='rain_mean').head(10)

# Create a vertical bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='ward_lgd_n', y='rain_mean', data=df_sorted)
plt.xticks(rotation=45)
plt.xlabel('Ward LGD N')
plt.ylabel('Rain Mean')
plt.title('Lowest 10 Wards based on Rain Mean')
plt.show()
```
